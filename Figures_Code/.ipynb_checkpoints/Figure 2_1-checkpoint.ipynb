```python
E = np.arange(0, 2.01, 0.01, dtype='float128')
J = np.arange(0, 8.01, 0.01, dtype='float128')
theta_arr = np.arange(0,5,1, dtype='float128')
alpha = np.arange(1, 1000.01, 0.01)
idx_alpha3 = np.where(np.abs(alpha.round(2) - 3.0) < 1e-4)[0]
alpha[idx_alpha3]
```

```python
J_cusp = np.zeros((len(alpha), len(theta_arr)))
E_cusp = np.zeros((len(alpha), len(theta_arr)))

for i in range(len(theta_arr)):
    theta = theta_arr[i]
    J_cusp[:,i] = theta + ((alpha - 1)/(alpha + 1))**((1-alpha)/alpha)
    E_cusp[:,i] = theta + ((alpha - 1)/(alpha + 1))**((1+alpha)/alpha)
```

```python
# Phase plot for alpha = 1
E_plot_a1 = np.arange(0, 2.001, 0.001)
J_plot_a1 = 2 + 2 * np.sqrt(1-E_plot_a1)
```

```python
# Phase plot for alpha = 2
E_plot_A2_Plus = np.zeros((len(J), len(theta_arr)), dtype='float128')
E_plot_A2_Minus = np.zeros((len(J), len(theta_arr)), dtype='float128')

for i in range(len(theta_arr)):
    theta = theta_arr[i]
    a2 = -27
    b2 = 18 * (J + 2*theta) * (2*J*theta + 1 + theta**2) - 4 * (J + 2*theta)**3 - 54 * J * theta**2
    c2 = 18 * J * theta**2 * (J + 2*theta) * (2*J*theta + 1 + theta**2) - 4 * J * theta**2 * (J + 2*theta)**3 + (J + 2*theta)**2 * (2*J*theta + 1 + theta**2)**2 - 4 * (2*J*theta + 1 + theta**2)**3 - 27 * J**2 * theta**4

    E_plot_A2_Plus[:,i] = (-b2 + np.sqrt(b2**2 - 4 * a2 * c2)) / (2 * a2)
    E_plot_A2_Minus[:,i] = (-b2 - np.sqrt(b2**2 - 4 * a2 * c2)) / (2 * a2)

    idx = np.where(J > theta)[0][0]
    E_plot_A2_Plus[:idx,i] = np.nan
    E_plot_A2_Minus[:idx,i] = np.nan
```

```python
# Phase plot for alpha = 3
E_plot_A3_1 = np.zeros((len(J), len(theta_arr)), dtype='float128')
E_plot_A3_2 = np.zeros((len(J), len(theta_arr)), dtype='float128')
E_plot_A3_3 = np.zeros((len(J), len(theta_arr)), dtype='float128')

# E_plot_A3_1 = mp.matrix(len(J), len(theta_arr))
# E_plot_A3_2 = mp.matrix(len(J), len(theta_arr))
# E_plot_A3_3 = mp.matrix(len(J), len(theta_arr))


for i in range(len(theta_arr)):
    theta = theta_arr[i]
    b = -(J + 3*theta)
    c = 3*(J*theta + theta**2)
    d = (1 - 3*J*theta**2 - theta**3)

    P3 = -192 * b * d - 128 * c**2 + 144 * b**2 * c - 27 * b**4
    Q3 = 144 * c * d**2 - 6 * b**2 * d**2 - 80 * b * c**2 * d + 16 * c**4 + 18 * b**3 * c * d - 4 * b**2 * c**3
    R3 = 18 * b * c * d**3 - 27 * d**4 - 4 * c**3 * d**2 - 4 * b**3 * d**3 + b**2 * c**2 * d**2
    
    A3 = -256
    B3 = 768 * J * theta**3 + P3
    C3 = -(768 * J**2 * theta**6 + 2 * P3 * J * theta**3 + Q3)
    D3 = 256 * J**3 * theta**9 + P3 * J**2 * theta**6 + Q3 * J * theta**3 + R3
    
    p = (3 * A3 * C3 - B3**2) / (3 * A3**2)
    q = (2 * B3**3 - 9 * A3 * B3 * C3 + 27 * A3**2 * D3) / (27 * A3**3)
    
    # E_plot_A3_1[:,i] = 2 * np.sqrt(-p / 3) * np.cos( (1/3) * np.arccos( ((3*q)/(2*p)) * np.sqrt(-3/p)) ) - (B3 / (3 * A3))
    # E_plot_A3_2[:,i] = 2 * np.sqrt(-p / 3) * np.cos( (1/3) * np.arccos( ((3*q)/(2*p)) * np.sqrt(-3/p) ) - (2 * np.pi * 1 / 3) ) - (B3 / (3 * A3))
    # E_plot_A3_3[:,i] = 2 * np.sqrt(-p / 3) * np.cos( (1/3) * np.arccos( ((3*q)/(2*p)) * np.sqrt(-3/p) ) - (2 * np.pi * 2 / 3) ) - (B3 / (3 * A3))

    for k in range(len(J)):
        E_plot_A3_1[k,i] = np.array([(2 * mp.sqrt(-p[k] / 3) * mp.cos( (1/3) * mp.acos( ((3*q[k])/(2*p[k])) * mp.sqrt(-3/p[k])).real ) - (B3[k] / (3 * A3))).real],dtype='float128')[0]
        E_plot_A3_2[k,i] = np.array([(2 * mp.sqrt(-p[k] / 3) * mp.cos( (1/3) * mp.acos( ((3*q[k])/(2*p[k])) * mp.sqrt(-3/p[k]).real ) - (2 * mp.pi * 1 / 3) ) - (B3[k] / (3 * A3))).real],dtype='float128')[0]
        E_plot_A3_3[k,i] = np.array([(2 * mp.sqrt(-p[k] / 3) * mp.cos( (1/3) * mp.acos( ((3*q[k])/(2*p[k])) * mp.sqrt(-3/p[k]).real ) - (2 * mp.pi * 2 / 3) ) - (B3[k] / (3 * A3))).real],dtype='float128')[0]

    #     E_plot_A3_1[k,i] = (2 * mp.sqrt(-p[k] / 3) * mp.cos( (1/3) * mp.acos( ((3*q[k])/(2*p[k])) * mp.sqrt(-3/p[k])) ) - (B3[k] / (3 * A3))).real
    #     E_plot_A3_2[k,i] = (2 * mp.sqrt(-p[k] / 3) * mp.cos( (1/3) * mp.acos( ((3*q[k])/(2*p[k])) * mp.sqrt(-3/p[k]) ) - (2 * mp.pi * 1 / 3) ) - (B3[k] / (3 * A3))).real
    #     E_plot_A3_3[k,i] = (2 * mp.sqrt(-p[k] / 3) * mp.cos( (1/3) * mp.acos( ((3*q[k])/(2*p[k])) * mp.sqrt(-3/p[k]) ) - (2 * mp.pi * 2 / 3) ) - (B3[k] / (3 * A3))).real

    idx0 = np.where(np.abs(J - J_cusp[idx_alpha3,i].round(2)) < 1e-8)[0][0]
    E_plot_A3_1[0:idx0,i] = np.nan
    E_plot_A3_2[0:idx0,i] = np.nan
    E_plot_A3_3[0:idx0,i] = np.nan
    try:
        idx1 = np.where(np.diff(E_plot_A3_2[:,i]) >= 0.0)[0][0]
        E_plot_A3_2[idx1+1:,i] = np.nan
    except:
        continue
    idx1 = np.where(np.diff(E_plot_A3_3[:,i]) < 0)[0][0]
    E_plot_A3_3[:idx1,i] = np.nan

    # # D0 = B3**2 - 3 * A3 * C3
    # # D1 = 2 * B3**3 - 9 * A3 * B3 * C3 + 27 * A3**2 * D3

    # # C00_Plus = np.cbrt( 0.5 * (D1 + np.sqrt(D1**2 - 4 * D0**3)) )
    # # C00_Minus = np.cbrt( 0.5 * (D1 - np.sqrt(D1**2 - 4 * D0**3)) )
    # # zeta = 0.5 * (-1 + )

    # C00_Plus = np.cbrt(-0.5 * q + np.sqrt(0.25 * q**2 + (p**3) / 27))
    # C00_Minus = np.cbrt(-0.5 * q - np.sqrt(0.25 * q**2 + (p**3) / 27))
    # E_plot_A3_2[:,i] = C00_Plus - (p / (3 * C00_Plus))
    # E_plot_A3_3[:,i] = C00_Minus - (p / (3 * C00_Minus))
```
